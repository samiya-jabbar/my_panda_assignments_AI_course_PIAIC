```python
import pandas as pd
a = pd.read_csv('states0.csv')
b = pd.read_csv('states1.csv')
c = pd.read_csv('states2.csv')
d = pd.read_csv('states3.csv')
e = pd.read_csv('states4.csv')
f = pd.read_csv('states6.csv')
g = pd.read_csv('states7.csv')
h = pd.read_csv('states8.csv')
i = pd.read_csv('states9.csv')
#print(a,b,c,d,e,f,g,h,i)
```

```python
final = pd.concat([a,b,c,d,e,f,g,h,i]).drop_duplicates(subset='State', keep='last').reset_index(drop=True) 
#print(final) 
```

```python
us_census=final.drop(['TotalPop','Black', 'White','Hispanic','Native', 'Asian', 'Pacific'], axis = 1) 
#print(us_census)
```

```python
# export all records to a new csv file 
us_census.to_csv('us_census.csv', index=False)
```

```python
def clean_currency(x):
    """ If the value is a string, then remove currency symbol and delimiters
    otherwise, the value is numeric and can be converted
    """
    if isinstance(x, str):
        return(x.replace('$', '').replace(',', ''))
    return(x)

us_census['Income'] = us_census['Income'].apply(clean_currency).astype('float')
```
